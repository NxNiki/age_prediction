```python
# plot figure to illustarte brain age prediction:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data_size = 300

# chronological_age = np.random.normal(loc = 15, scale = 7, size = data_size)
chronological_age = np.random.uniform(low = 5, high = 30, size = data_size)
true_brain_age = 1.2*chronological_age -.015*chronological_age**2 + np.random.normal(loc = 0, scale = 2, size = data_size)
predict_brain_age = true_brain_age + np.random.normal(loc = 0, scale = 3, size = data_size)

df = pd.DataFrame({'chornological age': chronological_age,
                   'True': true_brain_age,
                   'Estimated': predict_brain_age})
```

```python
# sns.set_style('white')
# sns.set(font_scale=2)

df_plot = pd.melt(df, id_vars = ['chornological age'], value_name = 'Brain age', var_name = 'Type')

g = sns.lmplot(x="chornological age", y="Brain age", hue = 'Type', 
           data=df_plot, order = 2, scatter_kws={'alpha':0.7}, legend_out = False, height = 10)

g.axes[0,0].axvline(x=22, ymin=11, ymax = 16, linewidth=20, color='r')

    
```
